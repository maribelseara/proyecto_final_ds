Explicación de las columnas.

'Unnamed:0': Replica el índice, se eliminará.

'Fecha extrac' : La fecha en que el anuncio fue scrapeado de la web.

'Enlace' : dirección web del registro.

'Marca' : Empresa fabricante del vehículo (Ej: Mercedes, Volvo, Citroen)

'Modelo' : Nombre que le ha dado el fabricante a ese tipo específico de vehículo (Ej: 911, Corsa, Saxo, León)

'Precio' : Precio por el cual el anunciante quiere vender el vehículo.

'Localización' : Enlace de google maps a la ubicación aproximada del vehículo

'Potencia' : Velocidad con la que un motor puede hacer un determinado trabajo. Medida en caballos. 

'Tipo de vendedor': Indica si se trata de un vendedor profesional o particular

'Categoría' : Tipo de vehículo según su forma (Ej. monovolumen, sedán, cabrio)

'Tipo de vehículo' : Indica el uso que se le ha dado al vehículo (Ej. KM0, Demostración, Ocasión, Clásica)

'puertas' : cantidad de puertas que posee el vehículo

'Versión del país' : Algunos modelos sacan diferentes versiones para diferentes países. En este dataset todos los valores son 'España'

'Garantía' : Meses de garantía que ofrece el vendedor. Según la legalidad española el mínimo es 6 meses en el caso de vendedor particular y 12 en el caso de vendedor profesional.

'Kilometraje' : distancia que el vehículo ha recorrido en km.

'Año' : Año de la primera matrícula del vehículo

'Tipo de cambio' : Si la caja de cambios es de accionamiento manual o automático

'Capacidad' : cantidad de combustible y aire que se puede combinar en el interior del motor en cada ciclo de funcionamiento. Medido en cm3

'Consumo de combustible' : corresponde al consumo de combustible en medida litros por cada 100 km en entornos mixtos (mezcla de urbano y extraurbano)

'Color exterior' : color de la carrocería del vehículo

'Color original' : color de la carrocería del vehículo cuando se construyó

'Tracción': indica qué ruedas reciben potencia del motor (Valores: Tracción delantera, Tracción trasera, Tracción a las cuatro ruedas)

'plazas': indica cuantos ocupantes puede transportar legalmente el vehículo

'Número de marchas' : cantidad de posiciones motrices con las que cuenta la caja de cambios de un vehículo.

'Número de cilindros': cantidad de piezas en el motor para quemar combustible y producir energía

'Peso': Masa del vehículo en kg

'Tipo de combustible' : tipo de material que usa el vehículo para obtener energía (Ej. Diésel, Gasolina, Eléctrico) 

'Mes' : Mes en número de la primera matriculación del vehículo

'Ciudad' : Municipio donde está actualmente el vehículo

'CP': Código postal de la ubicación actual del vehículo

'provincia' : división territorial de España donde se encuentra actualmente el vehículo

Habrá que consultar la mediana para tener una visión más realista.

En la potencia vemos que el mínimo es 1. Ningún vehículo tiene 1 caballo de potencia se trata de un error al entrar los datos. Habrá que poner a NaN los que sean menores a 20 CV. Intentarlos rellenar con las mismas marcas y modelos y descartar los que no puedan ser rellenados.

En la variable 'puertas' todo parece correcto. Hay coches de 2 puertas y también los hay de 7, no parece tener datos incorrectos.

En la variable 'Garantía' vemos que sí hay datos incorrectos, ya que el mínimo es 1, cuando como explicamos en la parte de limpieza el mínimo legal es 6 meses en el caso de particulares y 12 en el caso de profesionales. Habrá que arreglarlo.

En la variable 'Kilomatraje' vemos que hay valores a 0. Una de las categorías de vehículos es km0, habrá que comprobar que se trata de coches de esa categoría y no de errores.

La variable 'Año' parece tener los datos correctos dado que hay coches clásicos que bien pueden ser de 1930.

En la variable 'Capacidad', ya que ningún coche puede tener 1 cm3 de capacidad. Revisaremos o borraremos todos los que tengan este valor inferior a 100cm3

En la variable 'Consumo de combustible' también parece haber algunas anomalías. Revisaremos si es por ser vehículos eléctricos o si es algún error.

Para la variable 'plazas' comprobaremos si los que la tienen puesta a 1 realmente son vehículos monoplazas o es un error.

En la variable 'Número de marchas' corregiremos o descartaremos los valores de menos de 4 marchas, porque no hay vehículos con esas características.

Para el atributo 'Número de cilindros' ningún coche tiene menos de 3 cilindros, habrá que arreglarlo.

En la variable 'Peso' habrá que revisar los coches de menos de 1000 kg, que realmente estén correctos.

El resto de variables se han generado durante el proceso de limpieza y se ha comprobado allí que no haya errores.

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```

Vamos a arreglar los valores irregulares del atributo 'Potencia', primero igualaremos los valores inferiores a 20 a NaN e intentaremos rellenar los máximos posibles.

```python
df.loc[df['Potencia'] < 20, 'Potencia'] = None
df['Potencia'].isna().sum()
```

```python
df['Potencia'] = df.groupby(['Marca', 'Modelo'])['Potencia'].fillna(method='ffill')
df['Potencia'] = df.groupby(['Marca', 'Modelo'])['Potencia'].fillna(method='bfill')
try:
    df['Potencia'] = df.groupby(['Marca'])['Potencia'].transform(lambda x: x.fillna((x.mode()[0])))
except:
    pass
df['Potencia'].isna().sum()
```

Si quedase algún registro que no hubiéramos podido rellenar lo eliminaríamos 

```python
df=df[df['Potencia'].notna()]
```

Vamos a arreglar la variable Garantía para cumplir la legalidad española, 12 meses mínimo para vendedores profesionales, 6 meses mínimo para particulares

```python
def ajustar_garantía(fila):
    if fila['Garantía'] >= 12:
        return fila['Garantía']
    if fila['Tipo vendedor'] == 'Prof.':
        return 12
    return 6
```

```python
df['Garantía']=df.apply(ajustar_garantía, axis=1)
```

```python
df['Garantía'].describe()
```

Ahora vamos a ajustar la variable 'Kilometraje'

```python
def ajustar_kilometraje(fila):
    if fila['Kilometraje'] >=100:
        return fila['Kilometraje']
    if fila['Tipo de vehículo']=='KM0' or fila['Tipo de vehículo']=='Demostración':
        return fila['Kilometraje']
    return None
```

```python
df['Kilometraje']=df.apply(ajustar_kilometraje, axis=1)
```

```python
df['Kilometraje'].isna().sum()
```

Como ya hemos dicho este no es un campo que podamos imputar a la ligera, así que eliminaremos los registros

```python
df=df[df['Kilometraje'].notna()]
```

```python
df['Kilometraje'].describe()
```

Vemos que nos quedan coches con el kilometraje a 0 pero en esta ocasión ya estamos seguros de que todos pertenecen al tipo 'KM0' o 'Demostración'.

A continuación ajustaremos la 'Capacidad', asumiremos que cualquier valor por debajo de 100 cm3 se trata de un error, lo pondremos a NaN y luego intentaremos rellenarlo con la capacidad de coches de la misma 'Marca y 'Modelo' y los posteriormente los NaN que queden con la moda de esa 'Marca'

```python
df.loc[df['Capacidad'] < 100, 'Capacidad'] = None
df['Capacidad'].isna().sum()
```

```python
df['Capacidad'] = df.groupby(['Marca', 'Modelo'])['Capacidad'].fillna(method='ffill')
df['Capacidad'] = df.groupby(['Marca', 'Modelo'])['Capacidad'].fillna(method='bfill')
try:
    df['Capacidad'] = df.groupby(['Marca'])['Capacidad'].transform(lambda x: x.fillna((x.mode()[0])))
except:
    pass
df['Capacidad'].isna().sum()
```

Para el consumo de combustible asumiremos que si pone que gasta menos de 2 litros cada 100 km se trata de un error excepto si estamos hablando de un vehículo eléctrico.

```python
def ajustar_consumo(fila):
    if fila['Consumo de combustible']>=2 or fila['Tipo de combustible']=='Eléctrico':
        return fila['Consumo de combustible']
    return None
```

```python
df['Consumo de combustible']=df.apply(ajustar_consumo, axis=1)
df['Consumo de combustible'].isna().sum()
```

```python
df['Consumo de combustible'] = df.groupby(['Marca', 'Modelo'])['Consumo de combustible'].fillna(method='ffill')
df['Consumo de combustible'] = df.groupby(['Marca', 'Modelo'])['Consumo de combustible'].fillna(method='bfill')
try:
    df['Consumo de combustible'] = df.groupby(['Marca'])['Consumo de combustible'].transform(lambda x: x.fillna((x.mode()[0])))
except:
    pass
df['Consumo de combustible'].isna().sum()
```

```python
df.loc[df['plazas'] < 2, 'plazas'] = None
df['plazas'].isna().sum()
```

```python
df['plazas'] = df.groupby(['Marca', 'Modelo'])['plazas'].fillna(method='ffill')
df['plazas'] = df.groupby(['Marca', 'Modelo'])['plazas'].fillna(method='bfill')
try:
    df['plazas'] = df.groupby(['Marca'])['plazas'].transform(lambda x: x.fillna((x.mode()[0])))
except:
    pass
df['plazas'].isna().sum()
```

En el caso del 'Número de marchas' no existen vehículos con menos de 4 marchas, así que pondremos los valores inferiores a 4 a NaN e intentaremos rellenarlos.

```python
df.loc[df['Número de marchas'] < 4, 'Número de marchas'] = None
df['Número de marchas'].isna().sum()
```

```python
df['Número de marchas'] = df.groupby(['Marca', 'Modelo'])['Número de marchas'].fillna(method='ffill')
df['Número de marchas'] = df.groupby(['Marca', 'Modelo'])['Número de marchas'].fillna(method='bfill')
try:
    df['Número de marchas'] = df.groupby(['Marca'])['Número de marchas'].transform(lambda x: x.fillna((x.mode()[0])))
except:
    pass
df['Número de marchas'] = df['Número de marchas'].transform(lambda x: x.fillna((x.mode()[0])))
df['Número de marchas'].isna().sum()
```

Para la variable 'Número de cilindros' no existen vehículos con menos de 3 cilindros, así que pondremos los que tienen valores inferiores a 3 en NaN y los rellenaremos.

```python
df.loc[df['Número de cilindros'] < 3, 'Número de cilindros'] = None
df['Número de cilindros'].isna().sum()
```

```python
df['Número de cilindros'] = df.groupby(['Marca', 'Modelo'])['Número de cilindros'].fillna(method='ffill')
df['Número de cilindros'] = df.groupby(['Marca', 'Modelo'])['Número de cilindros'].fillna(method='bfill')
try:
    df['Número de cilindros'] = df.groupby(['Marca'])['Número de cilindros'].transform(lambda x: x.fillna((x.mode()[0])))
except:
    pass
df['Número de cilindros'] = df['Número de cilindros'].transform(lambda x: x.fillna((x.mode()[0])))
df['Número de cilindros'].isna().sum()
```

En cuanto al peso, según este artículo https://www.motorpasion.com/suv/pagar-aparcar-cuanto-pesado-sea-tu-coche-medida-que-algunas-ciudades-van-a-imponer-breve#:~:text=En%20realidad%20es%20una%20cuesti%C3%B3n,de%201.020%20kg%20en%201990 la media de peso de los coches actuales es de 1400 kg y de los de hace 20 años 1200kg. Así que asumiremos que cualquier coche que pese menos de 1000kg se trata de un error, lo pondremos a NaN y lo rellenaremos.

```python
df.loc[df['Peso'] < 1000, 'Peso'] = None
df['Peso'].isna().sum()
```

```python
df['Peso'] = df.groupby(['Marca', 'Modelo'])['Peso'].fillna(method='ffill')
df['Peso'] = df.groupby(['Marca', 'Modelo'])['Peso'].fillna(method='bfill')
try:
    df['Peso'] = df.groupby(['Marca'])['Peso'].transform(lambda x: x.fillna((x.mean())))
except:
    pass
df['Peso'] = df['Peso'].transform(lambda x: x.fillna((x.mean())))
df['Peso'].isna().sum()
```
